# __Magnetotransporte__

## __Anisotropía de forma__
Un material anisotrópico es el cuál no tiene las mismas propiedades , en este caso, de magnetización; en todas las direcciones. Su estudio es relevante para en estudio de Resonancia Ferromagnética (FEM); técnica que sirve para pruebas de la dinamica de la magnetización en películas delgadas. (Callori et al. 2020)
### __Anisotropía de forma__
La magnetización $M$ se ve afectada por la forma macroscópica del sólido. En general se forma un campo desmagnetizante. Supongamos el siguiente sistema:<br>

THIN_FILM_V1.png




De la imagen se tiene las expresiones para:
$$\textbf{M}=M\left[\begin{matrix}
sen(\theta_M)cos(\phi_M)\\
sen(\theta_M)sen(\phi_M)\\
cos(\theta_M)
\end{matrix}
\right] $$
$$ \textbf{H}=H\left[\begin{matrix}


sen(\theta_H)cos(\phi_H)\\
sen(\theta_H)sen(\phi_H)\\
cos(\theta_H)
\end{matrix}
\right] $$
La energía magnetostática se define como:
$$ E_{ms}=-\frac{1}{2}\int \mu_0 \mathbf{M}\cdot \mathbf{H_d} dV$$
donde $\mathbf{H}_d=-\overline{N}\mathbf{M}$, es el campo desmagnetizante y $\overline{N}$ es el tensor de segundo orden denominado __tensor desmagnetizante__.


El tensor para una geometría esférica tiene la expresión:
$$
\overline{N} =\left( \begin{matrix}
 \frac{1}{3}& 0 & 0 \\
0 & \frac{1}{3}& 0 \\
0 & 0 & \frac{1}{3}
\end{matrix}\right)
$$
entonces, la expresión de la energía es:
$$E_{ms}=\frac{1}{6}\mu_0 M^2 $$

En el caso de una superficie elipsoide de revolución con semiejes a, b y c (a||x, b||y, c||z). SI se tiene la condición $a=b\neq c$, la energía magnestostática queda expresada como:
$$E_{ms}=\frac{1}{2}\mu_0M^2(N_a\sin^2\theta_M\cos^2\phi_M+N_c\sin^2\theta_M\sin^2\phi_M + N_a \cos^2\theta_M)$$
Supongamos una película delgada ($a\thicksim b \gg c$), en este caso  el tensor desmagnetizante es:
$$
\overline{N}=\left(\begin{matrix}
0&0&0\\
0&0&0\\
0&0&1
\end{matrix}\right)
$$
por lo que la energía magnetostática es:
$$E_{ms}=\frac{1}{2}\mu_0M^2cos^2\theta_M $$

En este caso el plano fácil se encuentra en el plano de la película y, por tanto, la magnetización se encontrará contenida en dicho plano al haber un campo magnético externo.

### __Anisotropía magnetocristalina__
Está relacionada con la simetría cristalina de la muestra en estudio.
Se origina debido al __entrelazamiento entre orbitales electrónicos$^1$__ con la estructura cristalina del material (_quenching_ del momento angular)


Se puede obtener de la mecánica cuántica (primeros principios) pero es más óptimo, desde el punto vista práctico, usar un acercamiento fenomenológico; escribiendolo como una serie de potencias de las componentes angulares de la magnetización$^2$




En general se tendría (usando notación de Einstein para simplificar)
$$
E_c=E_0+b_i\alpha_i+b_{ij}\alpha_i\alpha_i+b_{ijk}\alpha_i\alpha_j\alpha_k+\mathcal{O}(\alpha^5)
$$

donde se considera que los términos de orden mayor a 4, $\mathcal{O}(\alpha^5)$, son muy pequeños y por tanto se pueden despreciar.
Se deben tener en cuenta dos consideraciones: La magnetización tiene simetría de inversión; es decir:
$$E(\mathbf{M})=E(-\mathbf M) $$
$$E(\alpha_i)=E(-\alpha_i) $$
Y, además, de considerar el caso particular de una película cuya simetría cristalina sea una cúbica, y por tanto $b_{ij}=0 $ para $i\neq j$, es decir, los términos cruzados son cero, teniendon cuenta que en el sistema cúbico los indices son indistinguibles y por tanto $b_{ii}=b_{jj},~\forall~ i,j$. Bajo estas consideraciones la energía de anisotropía cristalina estaría dado por la expresión
$$E_c^{cub}=E_0+b_{11}+b_{1111}(\alpha^4_1+\alpha_2^4+\alpha_3^4)+ 6b_{1122}(\alpha_1^2\alpha^2_2+\alpha_1^2\alpha_3^2+\alpha_2^2\alpha_3^2)$$
O, podemos reescribirlo. considerando $\alpha_1^2+\alpha_2^2+\alpha_3^2$=0(condición de  normalización) como:
$$E_c^{cub}=K_0+K_1(\alpha_1^2\alpha^2_2+\alpha_1^2\alpha_3^2+\alpha_2^2\alpha_3^2)+K_2(\alpha_1^2\alpha_2^2\alpha_3^2)+... $$
donde K_i son coeficiente que dependen de los coeficiente $b_i$.<br>



###  Anisotropía Uniaxial
Se refiere a la orientación o plano preferencial de la estructura cristalina que puede ser inducida mediante la aplicación de un campo magnético externo o de un ángulo entre el sustrato y el blanco.  
La anisotropia uniaxial se suele expresar como:
$$K_u=K_0+K_1\sin^2\theta_M+K_2\sin^4\theta_M +K_3 \sin^6\theta_M+... $$
Esta expansión en $\sin \theta$ corresponde  a la simentría de 180º (simetría de inversión espacial) para un sistema uniaxial con correcciones decrecientes de orden superior; es decir, podemos considerar los términos de orden mayor a 2 como correcciones lo suficientemente insignificantes con respecto a $K_1$. Entonces la energía uniaxial está dada por:
$$E_u=-K_u\frac{(\mathbf{M}\cdot \hat{e}_u)^2}{|\mathbf{M}|^2}$$
donde $K_u$ es la constante de anisotropía uniaxial y $\hat{e}_u$ es el vector unitario correspondiente a la dirección del eje fácil uniaxial.

La energía libre está dada por:
$$F=-\mathbf H\cdot \mathbf M  +\frac{1}{2}\mathbf M\cdot \bar{\mathbf N}\cdot \mathbf M-\frac{K_u}{M^2}(\hat{e}_u\cdot \mathbf{M})^2-K_c(\alpha_1^2\alpha_2^2+\alpha_2^2\alpha_3^2+\alpha_3^2\alpha_1^2)$$

En nuestro caso particular se va a usar un campo paralelo al plano, por tanto: $\theta_H=\pi/2$, obtenemos entonces:

$$F=-M H \sin(\theta_M) \cos(\phi_M - \phi_H) + 2 \pi M^2 \cos^2(\theta_M) - \frac{K_c}{4} \left( \sin^2(2\theta_M) + \sin^2(2\phi_M) \sin^4(\theta_M) \right) - K_u \sin^2(\theta_M) \cos^2(\phi_M - \phi_u)$$
Derivando con respecto a $\theta_M$:



$$
\frac{dF}{d\theta} = M H \cos(\theta_M) \cos(\phi_M - \phi_H) - 2 \pi M^2 \sin(2\theta_M)  - \frac{K_c}{2} \left( \sin(4\theta_M) + \sin^2(\theta_M) \sin(2\theta_M) \sin^2(2\phi_M) \right)+K_u \sin(2\theta_M) \cos^2(\phi_M - \phi_u)
$$
E igualamos esta expresión a 0 obtenemos que :


```python
# @title Función de energía libre y su derivada
def F(M,H, K_u, K_c, theta_M, phi_M,phi_H,phi_e):
    theta_M=np.radians(theta_M)
    phi_M=np.radians(phi_M)
    phi_H=np.radians(phi_H)
    phi_e=np.radians(phi_e)

    cos = lambda x: np.cos(x)
    sen = lambda x: np.sin(x)

    Zeeman=-M*H*sen(theta_M)*cos(phi_M-phi_H)
    Ani_forma=2*np.pi*M**2*cos(theta_M)**2
    Ani_mag_cris=-0.25*K_c*(sen(2*theta_M)**2+sen(2*phi_M)**2*sen(theta_M)**4)
    Ani_uniaxial=-K_u*sen(theta_M)**2*cos(phi_M-phi_e)**2
    return Zeeman+Ani_forma+Ani_uniaxial+Ani_mag_cris

def Fd(M, H, K_u, K_c, theta_M, phi_M,phi_H,phi_e):
    theta_M=np.radians(theta_M)
    phi_M=np.radians(phi_M)
    phi_H=np.radians(phi_H)
    phi_e=np.radians(phi_e)

    cos = lambda x: np.cos(x)
    sen = lambda x: np.sin(x)
    # Derivadas de cada término
    dZeeman_dtheta = M * H * cos(theta_M)*cos(phi_M-phi_H)
    dAni_forma_dtheta = -2*np.pi*M**2 *sen(2*theta_M)
    dAni_uniaxial_dtheta = K_u*sen(2*theta_M)*cos(phi_M-phi_e)**2
    dAni_mag_cris_dtheta = -0.5*K_c*(sen(4*theta_M)+sen(theta_M)**2*sen(2*theta_M)*sen(2*phi_M)**2)

    return dZeeman_dtheta + dAni_forma_dtheta + dAni_uniaxial_dtheta + dAni_mag_cris_dtheta
```

```python
# @title Gráfica de la función

# Valores iniciales y parámetros
M=170
H=45
K_u=12*M/2
K_c=290*M/2

phi_M =0
phi_H=10
phi_e=10
theta_M_guess = 100
df=lambda theta_M: Fd(M,H,K_u,K_c,theta_M,phi_M,phi_H,phi_e)
# ANALISEMOS LA GRÁFICA
x=np.linspace(0,360,100)
plt.plot(x,df(x))
plt.xlabel("$\\theta$ (rads)")
plt.ylabel(r"$ \frac{dF}{d\theta}$")
plt.title("Derivada de la energía libre")
plt.grid()
plt.show()
```

```python
# @title Minimización de la energía y determinación del punto crítico $\theta_M$
x0=root(df,theta_M_guess)
print(f"El valor de theta_M " + f"donde la energía se minimiza es :{x0.x[0]}º")
#Analisemos para varios valores de phi_H y phi_M
phi_H=np.linspace(0,360,10)
phi_M=np.linspace(0,360,10)
phi_u=np.linspace(0,360,10)
for i in phi_H:
  for j in phi_M:
    for k in phi_u:
      df=lambda theta_M: Fd(M,H,K_u,K_c,theta_M,j,i,k)
      x0=root(df,theta_M_guess)
      if abs(round(x0.x[0])-90)!=0:
        print(f"El valor de theta_M " + f"donde la energía se minimiza para phi_H={i}º, phi_M={j}º y phi_u={k}º  es :{x0.x[0]}º")
```

Tenemos que $\theta_M=90º$, entonces, la ecuación de la energía libre se reduce a:
$$F(\phi_M)=-M H\cos(\phi_M - \phi_H) - \frac{K_c}{4}  \sin^2(2\phi_M) - K_u\cos^2(\phi_M - \phi_u)$$
Es decir, la magnetización se encuentra dentro del plano para un campo paralelo al plano.

## Método de Newton Raphson (NRM)
El algoritmo de Newton-Raphson(NR) es un método para poder encontrar las raices de una función $f(x)$. Sin embargo, este método requiere de la derivada; la cual puede calcularse de manera analítica o de manera numérica. La fórmula de NR se deriva de la serie de Taylor de $f(x)$ sobre x:
$$
  f(x_{i+1})=f(x_i)+f'(x_i)(x_{i+1}-x_i)^2+\mathcal{O}(x_{i+1}-x_i)^2
$$
Luego, si suponemos que $x_{i+1}$ es una raíz de f y que $\mathcal{O}(z)$ se puede despreciar, tenemos:
$$ 0=f(x_i)+f'(x_i)(x_{i+1}-x_i)^2$$
Despejando $x_{i+1}$:
$$x_{i+1}=x_i-\frac{f(x_i)}{f'(x_i)} $$
Una forma de evitar tener que calcular de manera analítica la función es aproximar la derivada de la forma:
$$f'(x_i)≈ \frac{f(x_i+h)-f(x_{i})}{h}$$
donde h es un número lo suficientemente pequeño para que la aproximación sea válida.
## Método Simplex Downhill o Nelder-Mead

### Descripción
El método de Nelder-Mead es un algoritmo numérico utilizado para minimizar una función en un espacio de variables de dimensión arbitraria, sin necesidad de calcular gradientes. Es particularmente útil para funciones no diferenciables o con ruido.

El método emplea un enfoque geométrico, donde un **simplex** (un polígono de \(n+1\) vértices en un espacio de \(n\) dimensiones) se desplaza por el espacio de soluciones a través de transformaciones como reflexiones, expansiones, contracciones y reducciones(véase la figura 1).<br>

### Algoritmo

1. **Inicialización**:
   Se comienza con un simplex de \(n+1\) vértices. El simplex inicial se elige en torno a una estimación inicial del mínimo.

2. **Evaluación de vértices**:
   Se evalúa la función objetivo en cada vértice del simplex.

3. **Transformaciones del simplex**:
   Dependiendo de los valores de la función en los vértices, se aplican las siguientes transformaciones para mover el simplex:

   - **Reflexión**: El peor vértice del simplex es reemplazado por uno reflejado a través del centroide de los otros vértices.
   - **Expansión**: Si la reflexión mejora significativamente la función, el simplex se expande en esa dirección.
   - **Contracción**: Si la reflexión no mejora lo suficiente, el simplex se contrae hacia el mejor vértice.
   - **Reducción**: Si ninguna de las transformaciones anteriores mejora el resultado, todos los vértices se acercan al mejor vértice.

4. **Criterio de convergencia**:
   El proceso se repite hasta que el simplex converge, es decir, los vértices están suficientemente cerca unos de otros o los valores de la función en los vértices son casi iguales.

### Ventajas
- No requiere el cálculo de gradientes, por lo que es útil para funciones no derivables o con discontinuidades.
- Es relativamente sencillo de implementar.

### Desventajas
- Puede ser ineficiente para problemas de alta dimensión.
- Puede converger a mínimos locales en lugar del mínimo global.

```python
# @title
def Newthon_rapson(f,x0,dx,imax,errmin):
  xf=x0+dx
  do=True
  i=0
  while do:
    df=(f(xf)-f(x0))/(xf-x0)
    x0=xf
    xf=xf-f(xf)/df
    i=i+1
    if i>imax or abs(f(xf))<errmin:
      do=False
  return xf
def Down_the_hill(f, x0, alpha=0.1, imax=1e3, errmin=1e-10):
    xn = x0
    i = 0
    f_val = f(xn)

    while i < imax and abs(f_val) > errmin:
        # Calcular el paso basado en el valor de la función
        step = -alpha * f_val / abs(f_val) if f_val != 0 else -alpha

        # Actualizar el valor de xn
        xn += step

        # Recalcular f(xn)
        f_val = f(xn)
        i += 1

    return xn
def downhill(F, x0, side=0.1, imax=1e4, errmin=1e-6):
  n = len(x0)
  x = np.zeros((n+1,n))
  f = np.zeros(n+1)
  # Se genera el simplex inicial
  x[0] = x0

  for i in range(1,n+1):
    x[i]=x0
    x[i,i-1] = x0[i-1]+side
  # Elementos de los vertices
  for i in range(1,n+1):
    f[i]=F(x[i])

  # Bucle principal
  k=0
  do=True
  while do:
    # Encontrar los valores mas altos y bajos
    # de los vertices
    vlo=np.argmin(f)
    vhi=np.argmax(f)

    # Calculamos el de movimiento d
    d= (-(n+1)*x[vhi]+np.sum(x,axis=0))/n
    # revisamos la convergencia
    err=(np.sqrt(np.dot(d,d)/n))
    if (k>imax) or err<errmin:
      do=False
    else:
      do=True
    # reflexion
    xnew=x[vhi]+2.0*d
    fnew=F(xnew)
    if fnew<=f[vhi]:
      x[vhi]=xnew
      f[vhi]=fnew
    # expandir la reflexion
      xnew=x[vhi]+d
      fnew=F(xnew)
      if fnew<=f[vlo]:
        x[vlo]=xnew
        f[vlo]=fnew
    else:
      if fnew<=f[vhi]:
        xnew=x[vhi]+d
        fnew=F(xnew)
      else:
        xnew=x[vhi]+0.5*d
        fnew=F(xnew)
      if fnew<=f[vhi]:
        x[vhi]=xnew
        f[vhi]=fnew
      else:
        for i in range(len(x)):
          if i!=vlo:
            x[i]=x[i]+0.5*x[vlo]
            f[i]=F(x[i])
      k=k+1
  print(k+1)
  return x[vlo]
```
